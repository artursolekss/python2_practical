```python
from cassandra.cluster import Cluster
import uuid
import sqlite3
import time
import random
```

```python
uuid.uuid4()
```

```python
cluster = Cluster(["127.0.0.1"],port=9042)
```

```python
session = cluster.connect()
```

```python
create_keyspace_query = """
    CREATE KEYSPACE IF NOT EXISTS users_keyspace
    WITH replication = {'class':'SimpleStrategy', 'replication_factor':1}
"""

session.execute(create_keyspace_query)
```

```python
session.execute("USE users_keyspace")
session.execute(""" CREATE TABLE IF NOT EXISTS users (id UUID PRIMARY KEY, name TEXT, age INT ) """)
```

```python
insertion_query = """INSERT INTO users (id,name,age) VALUES (?,?,?)"""
```

```python
prepared_stmt = session.prepare(insertion_query)
```

```python
session.execute(prepared_stmt,(uuid.uuid4(),"Alice",30))
session.execute(prepared_stmt,(uuid.uuid4(),"Bob",28))
```

```python
result = session.execute("""SELECT * FROM users""")
for row in result:
    print(row.id,row.name,row.age)
```

```python
sqlite_conn = sqlite3.connect(":memory:")
```

```python
sqlite_cursor = sqlite_conn.cursor()
```

```python
sqlite_cursor.execute(""" CREATE TABLE users (id TEXT PRIMARY KEY, name TEXT, age INT )  """)
```

```python
num_records = 10000
data = [(uuid.uuid4(),f'User {i}',i % 100) for i in range(num_records)]
```

```python
#Cassandra

insertion_query = """INSERT INTO users (id,name,age) VALUES (?,?,?)"""
prepared_stmt = session.prepare(insertion_query)

start_time = time.time()

for record in data:
    session.execute_async(prepared_stmt,record)

end_time = time.time()
insertion_time_cassandra = end_time - start_time
```

```python
insertion_time_cassandra
```

```python
#SQL

data_sql = [(str(uuid.uuid4()),f'User {i}',i % 100) for i in range(num_records)]
insertion_query = """INSERT INTO users (id,name,age) VALUES (?,?,?)"""

start_time = time.time()

sqlite_cursor.executemany(insertion_query,data_sql)

end_time = time.time()
insertion_time_sql = end_time - start_time
```

```python
insertion_time_sql
```

```python
random_number = random.randint(0,num_records-1)
name = f'User {random_number}'
```

```python
#Cassandra

# selection_query = """SELECT id FROM users WHERE name = {}"""
# prepared_stmt = session.prepare(selection_query)

start_time = time.time()
result = session.execute("SELECT id FROM users WHERE name = '{}' ALLOW FILTERING".format(name))
end_time = time.time()
selection_cassandra = end_time - start_time
for row in result:
    print(row.id)

print(f'Cassandra time {selection_cassandra}')
```
