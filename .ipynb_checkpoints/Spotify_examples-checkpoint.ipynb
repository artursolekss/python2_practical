```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
```

```python
spotify = spotipy.Spotify(client_credentials_manager=SpotifyOAuth(redirect_uri='http://localhost:5000/callback',
                                                                  scope="user-modify-playback-state playlist-read-private"))
```

```python
artist_Dav_Bow = spotify.search(type='artist',q='David Bowie')
```

```python
Davie_Bowie_item = artist_Dav_Bow["artists"]["items"][0]
```

```python
Davie_Bowie_item
```

```python
Davie_Bowie_id = Davie_Bowie_item["id"]
```

```python
albums_Davie_Bowie = spotify.artist_albums(Davie_Bowie_id,limit=50)
```

```python
albums_Davie_Bowie_list = albums_Davie_Bowie["items"]
```

```python
for album in albums_Davie_Bowie_list:
    print(album["name"])
```

```python
test_one_album_id = albums_Davie_Bowie_list[0]["id"]
```

```python
test_one_album_id
```

```python
album_tracks = spotify.album_tracks(test_one_album_id,limit=50)
```

```python
album_tracks_items = album_tracks["items"]
```

```python
for song in album_tracks_items:
    print(song["name"])
```

```python
yesterday_search_result = spotify.search(q="Yesterday",type="track")
```

```python
yesterday_song_item = yesterday_search_result["tracks"]["items"][0]
```

```python
yesterday_song_id = yesterday_song_item["id"]
```

```python
album = yesterday_song_item["album"]
```

```python
album_id = album["id"]
```

```python
album_tracks = spotify.album_tracks(album_id,limit=50)
```

```python
for song in album_tracks["items"]:
    print(song["name"])
```

```python
Beatles_albums = spotify.artist_albums(Beatles_id,limit=50)
```

```python
for album in Beatles_albums["items"]:
    print(album["name"])
```
